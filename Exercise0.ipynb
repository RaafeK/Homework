```python
f = open('testdata.txt')
parsed_lines = []
for line in f:
    l = line.split(',') #create a list by splitting the string line at every ','
    l = [float(x) for x in l] #in-line for-loop that casts strings to floats
    parsed_lines.append(l)
    
plt.plot(parsed_lines[0])
```
